## Stocks/Shows/IMDB Datasets Processing and Combine

```python
import pandas as pd
```

### Import DataSets

```python
df_imdb = pd.read_csv("../data/cleaned_Netflix_IMDB.csv")
df_shows = pd.read_csv("../data/cleaned_Netflix_Shows.csv")
df_stocks = pd.read_csv("../data/cleaned_Netflix_stocks.csv")
```

```python
df_imdb.head()
```

```python
df_imdb['release_year'] = pd.to_datetime(df_imdb['release_year'], format='%Y', errors='coerce')
df_imdb['year_month'] = df_imdb['release_year'].dt.to_period('M')  
```

```python
df_imdb.info()
```

```python
df_shows.head()
```

```python
df_shows['date_added'] = pd.to_datetime(df_shows['date_added'], errors='coerce')
df_shows['year_month'] = df_shows['date_added'].dt.to_period('M') 
```

```python
df_shows.info()
```

```python
df_stocks.head()
```

```python
df_stocks['date'] = pd.to_datetime(df_stocks['date'])
df_stocks['year_month'] = df_stocks['date'].dt.to_period('M')  
```

```python
df_stocks.info()
```

### Combine Datasets

```python
df_shows.rename(columns={"date_added": "release_date"}, inplace=True)
```

```python
if "country" not in df_imdb.columns:
    df_imdb["country"] = "Unknown"

if "listed_in" in df_shows.columns:
    df_shows["genres"] = df_shows["listed_in"]
else:
    df_shows["genres"] = None

if "director" not in df_imdb.columns:
    df_imdb["director"] = pd.NA
```

```python
df_shows['type'] = df_shows['type'].str.upper().replace({
    "TV SHOW": "SHOW",
    "Movie": "MOVIE"
})

df_imdb['type'] = df_imdb['type'].str.upper().replace({
    "TV SHOW": "SHOW",
    "Movie": "MOVIE"
})
```

```python
df_shows["imdb_score"] = None
df_shows["imdb_votes"] = None
```

```python
df_shows["release_date"] = pd.to_datetime(df_shows["release_date"])
df_imdb["release_date"] = pd.to_datetime(df_imdb["release_date"])
```

```python
columns_to_keep = ["title", "type", "release_date", "imdb_score", "imdb_votes", 
                   "genres", "country", "director"]

df_shows = df_shows[columns_to_keep]
df_imdb = df_imdb[columns_to_keep]
```

```python
df_imdb = df_imdb.dropna(axis=1, how='all')
df_shows = df_shows.dropna(axis=1, how='all')

df_content = pd.concat([df_imdb, df_shows], ignore_index=True)
```

```python
df_content.head()
```

```python
df_content = df_content[df_content["release_date"] < "2022-01-01"]
```

```python
df_content["year_month"] = df_content["release_date"].dt.to_period("M")
```

```python
df_content["director"] = df_content["director"].fillna("Unknown")
```

```python
df_content_grouped = df_content.groupby(['title', 'year_month'], as_index=False).first()
```

```python
df_stocks = df_stocks.sort_values("date")
df_merged = pd.merge_asof(df_content_grouped.sort_values("release_date"), df_stocks, left_on="release_date", right_on="date", direction="nearest")
```

```python
df_merged.head()
```

```python
df_merged.info()
```

```python
df_merged.describe()
```

```python
df_merged.isnull().sum()
```

```python
df_merged['imdb_score'] = df_merged['imdb_score'].fillna(
    df_merged.groupby('type')['imdb_score'].transform('mean')
)

df_merged['imdb_votes'] = df_merged['imdb_votes'].fillna(
    df_merged.groupby('type')['imdb_votes'].transform('mean')
)
```

```python
df_merged = df_merged.dropna(subset=['daily_return', 'volatility', 'MA30'])
```

```python
df_merged.head()
```

```python
df_merged.to_csv('netflix_final_merged.csv', index=False)
```

### EDA

### General Analysis

```python
import plotly.express as px

fig = px.histogram(df_merged, x="imdb_score", color="type", 
                   marginal="box", nbins=30, opacity=0.7,
                   title="IMDb Score Distribution (Movies vs. TV Shows)")
fig.show()
```

```python
fig = px.scatter(df_merged, x="imdb_score", y="imdb_votes", 
                 color="type", size="imdb_votes", 
                 log_y=True, opacity=0.7,
                 title="IMDb Score vs. Number of Votes")
fig.show()
```

```python
df_content_per_month = df_merged.groupby("year_month_x")["title"].count().reset_index()
df_stock_volatility = df_merged.groupby("year_month_x")["volatility"].mean().reset_index()

df_content_per_month["year_month_x"] = df_content_per_month["year_month_x"].astype(str)
df_stock_volatility["year_month_x"] = df_stock_volatility["year_month_x"].astype(str)

fig = px.line(df_content_per_month, x="year_month_x", y="title", 
              title="Number of New Netflix Content Releases per Month",
              labels={"title": "Number of New Releases", "year_month_x": "Date"})
fig.show()

fig2 = px.line(df_stock_volatility, x="year_month_x", y="volatility", 
               title="Netflix Stock Volatility per Month",
               labels={"volatility": "Stock Volatility", "year_month_x": "Date"})
fig2.show()
```

```python
df_quarterly_content = df_merged.resample('QE', on='release_date').size().reset_index(name='new_content_count')

df_quarterly_stock = df_merged.resample('QE', on='date')['close'].mean().reset_index()

df_trend = df_quarterly_content.merge(df_quarterly_stock, left_on='release_date', right_on='date')

fig = px.line(df_trend, x='release_date', y=['new_content_count', 'close'],
              title="Quarterly Netflix Content Growth vs. Stock Price",
              labels={'new_content_count': 'New Content Count', 'close': 'Stock Price'})
fig.show()
```

```python
df_quarterly_ratings = df_merged.resample('QE', on='release_date')['imdb_score'].mean().reset_index(name='avg_imdb_score')

df_quarterly_volatility = df_merged.resample('QE', on='date')['volatility'].mean().reset_index(name='avg_volatility')

df_correlation = df_quarterly_ratings.merge(df_quarterly_volatility, left_on='release_date', right_on='date')

fig = px.scatter(df_correlation, x='avg_imdb_score', y='avg_volatility',
                 trendline="ols", title="Average IMDb Score vs. Stock Volatility",
                 labels={'avg_imdb_score': 'Avg IMDb Score', 'avg_volatility': 'Stock Volatility'})
fig.show()
```

```python
df_quarterly_stock = df_merged.resample('QE', on='date')['close'].mean().reset_index()

df_correlation = df_quarterly_ratings.merge(df_quarterly_volatility, left_on='release_date', right_on='date')
df_correlation = df_correlation.merge(df_quarterly_stock, on='date', how='left')
```

```python
df_correlation["price_change"] = df_correlation["close"].pct_change()

fig = px.scatter(df_correlation, x='avg_imdb_score', y='price_change',
                 trendline="ols", title="Average IMDb Score vs. Stock Price Change")
fig.show()
```

### Exploration: Does Content Quality Affect Netflix's Stock Price? 

```python
top_content = df_merged[(df_merged["imdb_score"] >= 8.5) & (df_merged["imdb_votes"] >= 100000)]

top_content_stock = df_merged[df_merged["title"].isin(top_content["title"])][["title", "release_date", "close"]].copy()

for days in [7, 30]:
    top_content_stock[f"price_change_{days}d"] = (
        top_content_stock["close"].pct_change(periods=days)
    )

fig = px.box(
    top_content_stock.melt(id_vars=["title"], value_vars=["price_change_7d", "price_change_30d"]),
    x="variable",
    y="value",
    title="Stock Price Change Before and After Major Netflix Releases",
    labels={"variable": "Time Period", "value": "Stock Price Change"}
)
fig.show()
```

```python
top_content["year_month"] = top_content["release_date"].astype(str).str[:7]

volatility_by_month = df_merged.groupby("year_month_y")["volatility"].mean().reset_index()

volatility_by_month["is_top_content_month"] = volatility_by_month["year_month_y"].isin(top_content["year_month_y"])

fig = px.box(
    volatility_by_month, x="is_top_content_month", y="volatility",
    title="Stock Volatility in Months with High-Rated Netflix Content",
    labels={"is_top_content_month": "High-Rated Content Released?", "volatility": "Stock Volatility"}
)
fig.show()
```

```python
df_early = df_merged[df_merged["release_date"] < "2010"]
df_recent = df_merged[df_merged["release_date"] > "2015"]

df_early_avg = df_early.resample("YE", on="release_date")[["imdb_score", "volatility"]].mean().reset_index()
df_recent_avg = df_recent.resample("YE", on="release_date")[["imdb_score", "volatility"]].mean().reset_index()

fig = px.line(df_early_avg, x="release_date", y=["imdb_score", "volatility"],
              title="Netflix Early Years: IMDb Score vs. Stock Volatility",
              labels={"imdb_score": "Average IMDb Score", "volatility": "Stock Volatility"})
fig.show()

fig2 = px.line(df_recent_avg, x="release_date", y=["imdb_score", "volatility"],
               title="Netflix Recent Years: IMDb Score vs. Stock Volatility",
               labels={"imdb_score": "Average IMDb Score", "volatility": "Stock Volatility"})
fig2.show()
```

```python
df_content_per_year = df_merged.groupby(df_merged["release_date"].dt.year)["title"].count().reset_index()
df_score_per_year = df_merged.groupby(df_merged["release_date"].dt.year)["imdb_score"].mean().reset_index()

df_trend = df_content_per_year.merge(df_score_per_year, on="release_date")

fig = px.line(df_trend, x="release_date", y=["title", "imdb_score"],
              title="Netflix Content Production vs. IMDb Score Over Time",
              labels={"title": "Number of Releases", "imdb_score": "Average IMDb Score"})
fig.show()
```

```python
df_merged["high_quality"] = df_merged["imdb_score"] > 8.0
df_high_quality_per_year = df_merged.groupby(df_merged["release_date"].dt.year)["high_quality"].mean().reset_index()

fig = px.line(df_high_quality_per_year, x="release_date", y="high_quality",
              title="Proportion of High-Quality Content (IMDb > 8.0) Over Time",
              labels={"high_quality": "Proportion of High IMDb Scores"})
fig.show()
```

```python
avg_rating_by_year = df_merged.groupby(df_merged["release_date"].dt.year)["imdb_score"].mean().reset_index()

fig = px.line(
    avg_rating_by_year, x="release_date", y="imdb_score",
    title="Average IMDb Score Trend Over Time",
    labels={"release_date": "Year", "imdb_score": "Average IMDb Score"}
)
fig.show()
```

```python
content_count_by_year = df_merged.groupby(df_merged["release_date"].dt.year)["title"].count().reset_index()

content_rating_trend = content_count_by_year.merge(avg_rating_by_year, on="release_date")

fig = px.scatter(
    content_rating_trend, x="title", y="imdb_score",
    title="Netflix Content Volume vs. IMDb Score",
    labels={"title": "Number of Releases", "imdb_score": "Average IMDb Score"},
    trendline="ols"
)
fig.show()
```

### Movie vs. TV show votes

```python
rating_trend_type = df_merged.groupby([df_merged["release_date"].dt.year, "type"])["imdb_score"].mean().reset_index()

fig = px.line(
    rating_trend_type, x="release_date", y="imdb_score", color="type",
    title="IMDb Score Trend: Movies vs. TV Shows",
    labels={"release_date": "Year", "imdb_score": "Average IMDb Score"}
)
fig.show()
```

```python
Q1 = df_merged["imdb_votes"].quantile(0.25)
Q3 = df_merged["imdb_votes"].quantile(0.75)
IQR = Q3 - Q1

df_filtered = df_merged[(df_merged["imdb_votes"] >= (Q1 - 1.5 * IQR)) & 
                        (df_merged["imdb_votes"] <= (Q3 + 1.5 * IQR))]

fig = px.box(
    df_filtered, x="type", y="imdb_votes",
    title="Filtered IMDb Votes Distribution: Movies vs. TV Shows",
    labels={"imdb_votes": "IMDb Votes", "type": "Content Type"},
    log_y=False
)
fig.show()
```

```python
content_trend = df_merged.groupby([df_merged["release_date"].dt.year, "type"])["title"].count().reset_index()

fig = px.line(content_trend, x="release_date", y="title", color="type",
              title="Netflix Content Production: TV Shows vs. Movies",
              labels={"release_date": "Year", "title": "Number of Releases", "type": "Content Type"})
fig.show()
```

### Netflix's content types vs. rating changes and country

```python
df_exploded_genres = df_merged.copy()
genre_trend = df_exploded_genres.groupby([df_exploded_genres["release_date"].dt.year, "genres"])["title"].count().reset_index()

genre_trend["total_per_year"] = genre_trend.groupby("release_date")["title"].transform("sum")
genre_trend["percentage"] = genre_trend["title"] / genre_trend["total_per_year"]

top_genres = genre_trend.groupby("genres")["title"].sum().nlargest(10).index
genre_trend_filtered = genre_trend[genre_trend["genres"].isin(top_genres)]

fig = px.area(
    genre_trend_filtered, x="release_date", y="percentage", color="genres",
    title="Netflix Content Genre Trends Over Time",
    labels={"release_date": "Year", "percentage": "Percentage of Content"}
)
fig.show()
```

```python
df_merged["is_international"] = df_merged["genres"].astype(str).str.contains("International")

df_merged["year"] = df_merged["release_date"].dt.year
international_trend = df_merged.groupby("year")["is_international"].mean().reset_index()

fig = px.line(
    international_trend, 
    x="year", y="is_international",
    title="Trend of International Content Over Time",
    labels={"year": "Year", "is_international": "Percentage of International Content"},
    markers=True
)
fig.show()
```

```python
import plotly.graph_objects as go

df_merged["year"] = pd.to_datetime(df_merged["release_date"]).dt.year

content_by_country = df_merged.groupby(["country", "year"]).size().reset_index(name="count")

recent_period = content_by_country[(content_by_country["year"] >= 2018) & (content_by_country["year"] <= 2021)]
past_period = content_by_country[(content_by_country["year"] >= 2010) & (content_by_country["year"] <= 2014)]

recent_total = recent_period.groupby("country")["count"].sum().reset_index(name="recent_count")
past_total = past_period.groupby("country")["count"].sum().reset_index(name="past_count")

df_growth = pd.merge(recent_total, past_total, on="country", how="left").fillna(0)

df_growth["growth_rate"] = (df_growth["recent_count"] - df_growth["past_count"]) / (df_growth["past_count"] + 1) * 100  

top_countries = df_growth.nlargest(11, "growth_rate")
```

```python
fig = go.Figure(data=go.Choropleth(
    locations=top_countries["country"],
    locationmode="country names",
    z=top_countries["growth_rate"],
    colorscale="Reds",
    colorbar_title="Growth Rate (%)",
    marker_line_color="darkgray",
    marker_line_width=0.5,
))

fig.update_layout(
    title_text="Top 11 Netflix Content Growth Rate by Country (2018-2021 vs. 2010-2014)",
    geo=dict(
        showframe=False,
        showcoastlines=True,
        projection_type="natural earth"
    )
)

fig.show()
```
